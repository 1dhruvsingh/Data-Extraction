```python
import matplotlib.pyplot as plt
import numpy as np

labels=['A','B','C']
data1=[3,2,5]
data2=[4,6,2]
data3=[5,3,4]

x=np.arange(len(labels))

plt.figure(figsize=(10,6))
plt.bar(x,data1,label='Data1')
plt.bar(x,data2,bottom=data1,label='Data2')
plt.bar(x,data3,bottom=np.array(data1)+np.array(data2),label='Data3')


plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.xticks(x,labels)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'D']
data1 = [3, 2, 5, 3]
data2 = [4, 6, 2 ,4]
data3 = [5, 3, 4 , 5]
data4 = [1, 7, 2 ,4]  # Add your new data here

x = np.arange(len(labels))  # Positions of the bars

plt.figure(figsize=(10, 6))

# Create the stacked bars with proper bottom positioning
bottom = np.zeros(len(x))  # Starting position for data1
plt.bar(x, data1, label='Data1', bottom=bottom)
bottom += data1  # Update bottom for data2
plt.bar(x, data2, label='Data2', bottom=bottom)
bottom += data2  # Update bottom for data3
plt.bar(x, data3, label='Data3', bottom=bottom)
bottom += data3  # Update bottom for data4 (if needed)
plt.bar(x, data4, label='Data4', bottom=bottom)  # Add the new bar

plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

set1=set(['A','B','C'])
set2=set(['B','C','D','E'])
set3=set(['D','E','C'])

plt.figure(figsize=(8,8))
venn3([set1,set2,set3],(set1,set2,set3))
plt.title('Venn Diagram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1= np.random.normal(0,1,100)
x2= np.random.normal(2,1,100)
y1= np.random.normal(0,1,100)
y2= np.random.normal(2,1,100)

plt.figure(figsize=(10,6))
plt.scatter(x1,y1,c='blue',label='Group1')
plt.scatter(x2,y2,c='red',label='Group2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dates= pd.date_range(start='2024-01-01', periods =10, freq='D')
values1= np.random.rand(10)
values2= np.random.rand(10)
values3= np.random.rand(10)

plt.figure(figsize=(10,6))
plt.plot(dates,values1,marker='o',label='Series1')
plt.plot(dates,values2,marker='x',label='Series2')
plt.plot(dates,values3,marker='o',label='Series3')
plt.title('Line Graph')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dates= pd.date_range(start='2024-01-01', periods =10, freq='D')
data1= np.random.rand(10)
data2= np.random.rand(10)
data3= np.random.rand(10)

plt.figure(figsize=(10,6))
plt.stackplot(dates, data1, data2, data3, labels=['Data1','Data2','Data3'])
plt.title('Stacked Area Chart')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

tasks=pd.DataFrame({
    'Task':['Task1','Task2','Task3'],
    'Start':['2024-01-01','2024-01-15','2024-01-20'],
    'End':['2024-01-10','2024-01-18','2024-01-25']
})

fig= px.timeline(tasks,x_start='Start',x_end='End',y='Task',title='Gantt Chart')
fig.update_yaxes(categoryorder='total ascending')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dates=pd.date_range(start='2023-01-01',periods=10,freq='D')
categories = ['A','B','C','D']
data = np.random.rand(10,4)
df=pd.DataFrame(data,columns=categories,index=dates)

plt.figure(figsize=(10,6))
sns.heatmap(df,cmap='YlGnBu',annot=True)
plt.title('Heatmap')
plt.xlabel('Category')
plt.ylabel('Date')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dates= pd.date_range(start='2024-01-01',periods=10,freq='D')
values=np.sin(np.linspace(0,2*np.pi,10))

plt.figure(figsize=(10,6))
plt.plot(dates,values,marker='o')
plt.title('Time Series Sequence')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show
```
