```python
import matplotlib.pyplot as plt
labels=['A','B','C','D']
sizes=[20,30,40,10]

fig, ax=plt.subplots()
wedges, texts ,autotexts=ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgesprops=dict(width=0.3))

centre_cicle = plt.Circle((0,0),0.2 ,color= 'white', fc='white')
fig.gca().add_artist(centre_cicle)

ax.axis('equal')
plt.title('Ring Chart')
plt.show()
```

```python
import plotly.graph_objects as go

lables=['root','A','B','A1','A2','B1']
parents = ['','root','root','A','A','B']

fig= go.Figure(go.Sunburst(
    labels=labels,
    parents=parents,
    values=[10,20,30,10,10,10]
))
fig.update_layout(margin=dict(t=0,l=0,r=0,b=0))
fig.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()
edges= [('Root','A'),('Root','B'),('A','A1'),('A','A2'),('B','B1')]
G.add_edges_from(edges)
pos= nx.spring_layout(G,seed=42)
plt.figure(figsize=(10,10))
nx.draw(G,pos,with_labels=True,arrows=True,node_size=2000,node_color='skyblue',font_size=10,font_weight='bold')
plt.title('tree diagram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
data1= np.random.normal(0,1,1000)
data2= np.random.normal(0,1.5,1000)

plt.figure(figsize=(10,6))
plt.hist(data1,bins=30, alpha=0.5, label='data1')
plt.hist(data2,bins=30, alpha=0.5, label='data2')
plt.title('Multiple Histograms')
plt.xlabel('Value')2
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
