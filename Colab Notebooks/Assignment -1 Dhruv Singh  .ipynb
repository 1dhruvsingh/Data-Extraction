```python
import pandas as pd

dftemp = pd.read_excel( "Results.xlsx",header=3)

subjects = {
'Result': "Statistics and Probability",
'Result.1': "Microsoft Excel",
'Result.2': "VBA",
'Result.3': "SQL",
'Result.4': "Power BI" }

df=dftemp.dropna()
results = {}

for index, row in df.iterrows():
  passed = []
  failed = []
  name=row["Student Name"]
  reg= int(row["Registration Number"])
  prog= row["Program Name"]
  date= pd.to_datetime( row['Exam Date']
  ).strftime('%d-%b-%y')

total = int(row["Grand Total"])

for col, subject in subjects.items():
  if row[col]== "Fail":
    failed.append(subject)
  else:
    passed.append(subject)

results[name] = {"Passed": passed, "Failed": failed}

passed_str=",".join(passed)
failed_str=",".join(failed)

with open(
  f"{name}.txt", "w+"
) as f:
 f.write(
     f"Mr. {name}, bearing registration number {reg},"
     f"enrolled in the program (prog), has appeared in the final exams"
      f"on {date}. The student has successfully qualified the following courses: "
      f" {passed_str} and has scored a grand total of {total} marks.")
if failed:
 f.write(
  f"\nHowever, the student has failed the following course(s): "
  f"{failed_str}. They may have to re-appear in this/these exam(s)." )
```
